```python
!pip install vosk sounddevice scipy numpy requests -q

import os, zipfile, requests, wave, json
import sounddevice as sd
from scipy.io.wavfile import write
from vosk import Model, KaldiRecognizer

MODEL_NAME = "vosk-model-ar-mgb2-0.4"
MODEL_ZIP = MODEL_NAME + ".zip"
MODEL_URL = f"https://alphacephei.com/vosk/models/{MODEL_ZIP}"

def ensure_vosk_model():
    if os.path.isdir(MODEL_NAME):
        print("‚úÖ ÿßŸÑŸÖŸàÿØŸäŸÑ ŸÖŸàÿ¨ŸàÿØ ÿ¨ÿßŸáÿ≤.")
        return MODEL_NAME
    print("‚¨áÔ∏è ÿ™ŸÜÿ≤ŸäŸÑ ÿßŸÑŸÖŸàÿØŸäŸÑ ÿßŸÑÿπÿ±ÿ®Ÿä...")
    with requests.get(MODEL_URL, stream=True) as r:
        r.raise_for_status()
        with open(MODEL_ZIP, "wb") as f:
            for ch in r.iter_content(8192):
                f.write(ch)
    print("üì¶ ŸÅŸÉ ÿßŸÑÿ∂ÿ∫ÿ∑...")
    with zipfile.ZipFile(MODEL_ZIP, "r") as z:
        z.extractall(".")
    os.remove(MODEL_ZIP)
    print("‚úÖ ÿ™ŸÖ ÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑŸÖŸàÿØŸäŸÑ:", MODEL_NAME)
    return MODEL_NAME

model_dir = ensure_vosk_model()

SAMPLE_RATE = 16000
DURATION = 5
WAV_PATH = "input.wav"

print("üéôÔ∏è ÿ™ŸÉŸÑŸÖŸä ÿßŸÑÿ¢ŸÜ...")
rec = sd.rec(int(DURATION * SAMPLE_RATE), samplerate=SAMPLE_RATE, channels=1, dtype='int16')
sd.wait()
write(WAV_PATH, SAMPLE_RATE, rec)
print("‚úÖ ÿ™ŸÖ ÿßŸÑÿ™ÿ≥ÿ¨ŸäŸÑ:", os.path.abspath(WAV_PATH))

print("üîç ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿµŸàÿ™ ÿ•ŸÑŸâ ŸÜÿµ...")
model = Model(model_dir)
recognizer = KaldiRecognizer(model, SAMPLE_RATE)

with wave.open(WAV_PATH, "rb") as wf:
    while True:
        data = wf.readframes(4000)
        if len(data) == 0:
            break
        recognizer.AcceptWaveform(data)

result_text = json.loads(recognizer.Result()).get("text", "").strip()
print("üó£Ô∏è ÿßŸÑŸÜÿµ:", result_text)
```

```python
# ================== ÿßŸÑÿ™ÿ´ÿ®Ÿäÿ™ (ÿ¥ÿ∫ŸëŸÑŸäŸáÿß ŸÖÿ±ÿ© ŸÅŸÇÿ∑ ÿ´ŸÖ ÿπŸÑŸëŸÇŸäŸáÿß) ==================
# !pip install -q ipywidgets vosk sounddevice scipy numpy requests cohere gTTS
# !jupyter nbextension enable --py widgetsnbextension

import os, sys, json, wave, time, zipfile, tempfile, requests, traceback
import sounddevice as sd
from scipy.io.wavfile import write
from vosk import Model, KaldiRecognizer
from gtts import gTTS
from IPython.display import display, HTML, Audio
import ipywidgets as widgets

# ================== ŸÖŸÅÿ™ÿßÿ≠ Cohere ==================
COHERE_API_KEY = "DeAgWFFRPhMQcgUdghEilKQzNrJfhqNruL17LUlw"

import cohere
co = cohere.Client(COHERE_API_KEY)

# ================== ÿ™ŸÜÿ≤ŸäŸÑ ŸÖŸàÿØŸäŸÑ Vosk ÿßŸÑÿπÿ±ÿ®Ÿä ==================
MODEL_NAME = "vosk-model-ar-mgb2-0.4"
MODEL_ZIP  = MODEL_NAME + ".zip"
MODEL_URL  = f"https://alphacephei.com/vosk/models/{MODEL_ZIP}"

def ensure_vosk_model():
    if os.path.isdir(MODEL_NAME):
        return MODEL_NAME
    print("‚¨áÔ∏è ÿ™ŸÜÿ≤ŸäŸÑ ŸÖŸàÿØŸäŸÑ Vosk ÿßŸÑÿπÿ±ÿ®Ÿä (ŸÖÿ±ÿ© Ÿàÿßÿ≠ÿØÿ©)...")
    with requests.get(MODEL_URL, stream=True, timeout=180) as r:
        r.raise_for_status()
        with open(MODEL_ZIP, "wb") as f:
            for ch in r.iter_content(1024 * 64):
                if ch: f.write(ch)
    print("üì¶ ŸÅŸÉ ÿßŸÑÿ∂ÿ∫ÿ∑...")
    with zipfile.ZipFile(MODEL_ZIP, "r") as z:
        z.extractall(".")
    os.remove(MODEL_ZIP)
    print("‚úÖ ÿ™ŸÖ ÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑŸÖŸàÿØŸäŸÑ.")
    return MODEL_NAME

model_dir = ensure_vosk_model()
vosk_model = Model(model_dir)

# ================== ÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ™ÿ≥ÿ¨ŸäŸÑ ==================
SAMPLE_RATE = 16000
TMP_WAV    = "input.wav"

def record_audio(seconds: int, device=None):
    sd.default.samplerate = SAMPLE_RATE
    if device is not None:
        sd.default.device = device
    rec = sd.rec(int(seconds * SAMPLE_RATE), samplerate=SAMPLE_RATE, channels=1, dtype='int16')
    sd.wait()
    write(TMP_WAV, SAMPLE_RATE, rec)
    return TMP_WAV

def stt_vosk(wav_path: str):
    rec = KaldiRecognizer(vosk_model, SAMPLE_RATE)
    with wave.open(wav_path, "rb") as wf:
        while True:
            data = wf.readframes(4000)
            if not data: break
            rec.AcceptWaveform(data)
    result = json.loads(rec.Result() or "{}")
    return (result.get("text") or "").strip()

def cohere_reply_ar(text: str) -> str:
    text = (text or "").strip()
    if not text:
        return "ŸÑŸÖ ÿ£ŸÑÿ™ŸÇÿ∑ ŸÉŸÑÿßŸÖŸãÿß Ÿàÿßÿ∂ÿ≠Ÿãÿß. ÿ¨ÿ±Ÿëÿ®Ÿä ŸÖÿ±ÿ© ÿ£ÿÆÿ±Ÿâ."
    try:
        r = co.generate(
            model="command",
            prompt=f"ÿ£ÿ¨ÿ® ÿ®ÿßŸÑŸÑÿ∫ÿ© ÿßŸÑÿπÿ±ÿ®Ÿäÿ© ÿ®ÿßÿÆÿ™ÿµÿßÿ± ŸàŸàÿ∂Ÿàÿ≠:\n{text}",
            max_tokens=200,
            temperature=0.6,
        )
        return (r.generations[0].text or "").strip() or "..."
    except Exception as e:
        return f"ÿ™ÿπÿ∞Ÿëÿ± ÿ™ŸàŸÑŸäÿØ ÿ±ÿØ: {e}"

def speak_ar_gtts(text: str):
    fn = os.path.join(tempfile.gettempdir(), f"reply_{int(time.time())}.mp3")
    gTTS(text=text, lang='ar').save(fn)
    display(Audio(filename=fn, autoplay=True))

# ================== Ÿàÿßÿ¨Ÿáÿ© ÿßŸÑŸÖÿ≠ÿßÿØÿ´ÿ© ==================
def render_chat(chat_log):
    css = """
    <style>
    .chat{font-family:'Segoe UI',Tahoma,Arial;max-width:760px}
    .msg{margin:6px 0;display:flex}
    .you{justify-content:flex-end}
    .bot{justify-content:flex-start}
    .b{display:inline-block;padding:10px 14px;border-radius:16px;max-width:85%;
       line-height:1.6;white-space:pre-wrap;word-wrap:break-word;box-shadow:0 1px 3px rgba(0,0,0,.06)}
    .you .b{background:#d1f4ff;border:1px solid #bde9f6}
    .bot .b{background:#f3f4f6;border:1px solid #e5e7eb}
    </style>
    """
    html = [css, '<div class="chat">']
    for m in chat_log:
        side = "you" if m["role"]=="user" else "bot"
        html.append(f'<div class="msg {side}"><div class="b">{m["text"]}</div></div>')
    html.append("</div>")
    return HTML("".join(html))

# ÿπŸÜÿßÿµÿ± ÿßŸÑÿ™ÿ≠ŸÉŸÖ
btn = widgets.Button(description="üé§ ÿ≥ÿ¨ŸëŸÑ 5 ÿ´ŸàÿßŸÜŸç", button_style="primary")
out = widgets.Output()
display(btn, out)
chat_log = []

def on_click(_):
    btn.disabled = True
    with out:
        out.clear_output()
        print("üéôÔ∏è ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ÿ≥ÿ¨ŸäŸÑ... ÿ™ŸÉŸÑŸÖŸä ÿßŸÑÿ¢ŸÜ")
        wav = record_audio(5)
        print("üîç ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿµŸàÿ™ ÿ•ŸÑŸâ ŸÜÿµ...")
        user_text = stt_vosk(wav) or "ŸÖÿß ŸÅŸáŸÖÿ™ ŸÉŸÑÿßŸÖŸÉ"
        chat_log.append({"role":"user","text":user_text})
        display(render_chat(chat_log))

        print("üí≠ ÿ™ŸàŸÑŸäÿØ ÿ±ÿØ ÿπÿ±ÿ®Ÿä...")
        reply = cohere_reply_ar(user_text)
        chat_log.append({"role":"assistant","text":reply})
        display(render_chat(chat_log))

        print("üîä ÿ™ÿ¥ÿ∫ŸäŸÑ ÿßŸÑÿ±ÿØ...")
        speak_ar_gtts(reply)
    btn.disabled = False

btn.on_click(on_click)
```
